```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from scipy.stats import linregress
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
X_train = train['x']
y_train = train['y']
```

```python
X_train, X_dropout, y_train, y_dropout = \
    train_test_split(X_train, y_train, test_size=0.1, random_state=19)
```

```python
plt.rcParams['figure.figsize'] = (8,5)
plt.scatter(X_train, y_train, c='blue', alpha=.3);
plt.title('Plot of X_train and y_train');
plt.xlabel('X_train');
plt.ylabel('y_train');
```

```python
Regress = LinearRegression()  
```

```python
Regress.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1));
```

```python
Prediction = Regress.predict(test['x'].values.reshape(-1,1))
```

```python
plt.rcParams['figure.figsize'] = (10,7)
plt.scatter(test['x'], test['y'], c='skyblue');
plt.title('Predicted Line v.s Values');
plt.xlabel('X_test');
plt.ylabel('y_test');
plt.plot(test['x'], Prediction, c='blue');
```
